```python
import collections
import os
import shutil
import pandas as pd
import math
```

```python
data_dir = "../data/cifar10/"
```

```python
def copyfile(filename, target_dir):
    """将文件复制到目标目录"""
    os.makedirs(target_dir, exist_ok=True)
    shutil.copy(filename, target_dir)
```

```python
def reorg_test():
    for item_name in os.listdir(os.path.join(data_dir, 'test')):
        item_path = os.path.join(data_dir, 'test',item_name )
        if os.path.isdir(item_path) == True:
            print("dir = " + item_name)
            for item_name_02 in os.listdir(item_path):
                fname = os.path.join(item_path,item_name_02)
                copyfile(fname, os.path.join(data_dir, 'train_valid_test', 'test','unknown'))

        else:
            print("filename =" + item_name )

#reorg_test()
```

```python
#@save
def rename_file():
    file_dir =  "../data/cifar10/train_valid_test/test/unknown"

    for file_name in os.listdir(file_dir):
        print(file_name.split('_')[0])
        print(file_name.split('.')[-1])
        new_name = file_name.split('_')[0] + '.' + file_name.split('.')[-1]
        print(new_name)
        prev_file = os.path.join(file_dir,file_name)
        new_file = os.path.join(file_dir,new_name)
        os.rename(prev_file, new_file)

        
#rename_file()
```

```python
def reorg_train_valid(valid_ratio):
    dir_name = "../data/cifar10/train_valid_test"

    for folder_name in os.listdir(os.path.join(dir_name, "train_valid")):
        file_counter = 0
        for file_name in os.listdir(os.path.join(dir_name, "train_valid", folder_name)):
            file_counter +=1
            fname = os.path.join(data_dir, 'train_valid', 'folder_name','file_name')
            if (math.floor(file_counter*valid_ratio) < 1) :  # 放入train 目录 
                copyfile(fname, os.path.join(data_dir, 'train_valid_test', 'train','folder_name'))
            else:
                file_counter = 0
                copyfile(fname, os.path.join(data_dir, 'train_valid_test', 'valid', 'folder_name'))
                

valid_ratio = 0.1
reorg_train_valid(valid_ratio)
```
